```python
# -*- coding: utf-8 -*-
"""
Created on Wed Mar 19 13:16:22 2025

@author: user
"""

import pandas as pd
import librosa
import soundfile as sf
from concurrent.futures import ThreadPoolExecutor
import torch
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2ForSequenceClassification
import random
import numpy as np
import openpyxl
import os
```

```python


def split_audio(audio, sr=16000, segment_length=60):
    """Splits audio into smaller segments of fixed length."""
    segment_samples = sr * segment_length
    return [audio[i:i+segment_samples] for i in range(0, len(audio), segment_samples) if len(audio[i:i+segment_samples]) >= sr]

# Load timestamps
timestamps = pd.read_excel("/content/drive/MyDrive/timestamps_clean.xlsx")

# Load feature extractor and model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"✅ Using device: {device}")

#feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("r-f/wav2vec-english-speech-emotion-recognition")
#model = Wav2Vec2ForSequenceClassification.from_pretrained("r-f/wav2vec-english-speech-emotion-recognition").to(device)
#feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("shivanshu292001/Emotions")
#model =  Wav2Vec2ForSequenceClassification.from_pretrained("shivanshu292001/Emotions").to(device)
#feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition")
#model =  Wav2Vec2ForSequenceClassification.from_pretrained("ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition").to(device)




#feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition")
#model =  Wav2Vec2ForSequenceClassification.from_pretrained("ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition").to(device)


feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("Dpngtm/wav2vec2-emotion-recognition")
model =  Wav2Vec2ForSequenceClassification.from_pretrained("Dpngtm/wav2vec2-emotion-recognition").to(device)

feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("DunnBC22/wav2vec2-base-Toronto_emotional_speech_set")
model =  Wav2Vec2ForSequenceClassification.from_pretrained("DunnBC22/wav2vec2-base-Toronto_emotional_speech_set").to(device)

feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("superb/wav2vec2-base-superb-er")
model =  Wav2Vec2ForSequenceClassification.from_pretrained("superb/wav2vec2-base-superb-er").to(device)

def predict_emotion_batch(audio_batch):
    inputs = feature_extractor(audio_batch, sampling_rate=16000, return_tensors="pt", padding=True)
    inputs = {key: val.to(device) for key, val in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    top3_indices = torch.argsort(predictions, dim=-1, descending=True)[:, :3]
    top3_probs = torch.gather(predictions, 1, top3_indices)

    emotions_top3 = [
        [
            (model.config.id2label[idx.item()], prob.item())
            for idx, prob in zip(indices, probs)
        ]
        for indices, probs in zip(top3_indices, top3_probs)
    ]

    return emotions_top3

def process_audio_batch(batch_ids):
    results = []

    for file_id in batch_ids:
        file_path = f"/content/drive/MyDrive/fomc_audio_cut/{file_id}.wav"
        if not os.path.exists(file_path):
            print(f"❌ File {file_path} not found!")
            continue

        audio, _ = librosa.load(file_path, sr=16000, mono=True)
        audio = librosa.util.normalize(audio)
        segments = split_audio(audio)

        if not segments:
            print(f"⚠️ Skipping {file_id} due to short audio.")
            continue

        segment_emotions = predict_emotion_batch(segments)

        avg_probs = {}
        for emotions in segment_emotions:
            for emotion, prob in emotions:
                avg_probs[emotion] = avg_probs.get(emotion, 0) + prob

        for emotion in avg_probs:
            avg_probs[emotion] /= len(segment_emotions)

        top3 = sorted(avg_probs.items(), key=lambda x: x[1], reverse=True)[:3]
        print(f"✅ {file_id}: {top3[0][0]} ({top3[0][1]:.2f}), {top3[1][0]} ({top3[1][1]:.2f}), {top3[2][0]} ({top3[2][1]:.2f})")
        results.append((file_id, *sum(top3, ())))

    return results

seeds = [42]
batch_size = 2

for seed in seeds:
    print(f"\n🚀 Running emotion detection with seed {seed}...\n")

    np.random.seed(seed)
    torch.manual_seed(seed)

    results = []
    with ThreadPoolExecutor(max_workers=4) as executor:
        for batch_ids in [timestamps["id"].tolist()[i:i + batch_size] for i in range(0, len(timestamps), batch_size)]:
            batch_result = executor.submit(process_audio_batch, batch_ids)
            if batch_result.result():
                results.extend(batch_result.result())

    if results:
        results_df = pd.DataFrame(results, columns=['id', 'emotion1', 'prob1', 'emotion2', 'prob2', 'emotion3', 'prob3'])
        timestamps_with_emotions = timestamps.merge(results_df, on="id", how="left")

        file_path = f"/content/drive/MyDrive/emotion_results_superb.csv"
        timestamps_with_emotions.to_csv(file_path, index=False)
        print(f"✅ Results saved: {file_path}")

print("\n🎉 All emotion detection runs completed successfully!")
```
